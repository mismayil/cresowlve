```python
from cresowlve.utils import read_json

model_results_path = "../experiments/outputs/gemini-3-flash-preview/chgk_en_benchmark_eval_reasoning_model_en_s0_gemini-3-flash-preview_20260302_151010.json"
benchmark_path = "../experiments/data/task/chgk_en_benchmark.json"
kg_source_path = "../experiments/data/task/chgk_benchmark_knowledge_source.json"
model_results = read_json(model_results_path)
benchmark = read_json(benchmark_path)
kg_source_results = read_json(kg_source_path)
```

```python
from collections import Counter
kg_sources = Counter([s["knowledge_source"] for s in kg_source_results["data"]])

kg_sources
```

```python
from statistics import mean
import random

russian_ids = [s["id"] for s in kg_source_results["data"] if s["knowledge_source"] == "russian"]
non_russian_ids = [s["id"] for s in kg_source_results["data"] if s["knowledge_source"] == "other"]
both_ids = [s["id"] for s in kg_source_results["data"] if s["knowledge_source"] == "both"]
russian_results = [r for r in model_results["data"] if r["id"] in russian_ids]
non_russian_results = [r for r in model_results["data"] if r["id"] in non_russian_ids]
both_results = [r for r in model_results["data"] if r["id"] in both_ids]

min_len = min(len(russian_results), len(non_russian_results), len(both_results))
russian_results = random.sample(russian_results, min_len)
non_russian_results = random.sample(non_russian_results, min_len)
both_results = random.sample(both_results, min_len)

russian_acc = mean([int(r.get("gpt-4o_judge_match", False)) for r in russian_results])
non_russian_acc = mean([int(r.get("gpt-4o_judge_match", False)) for r in non_russian_results])
both_acc = mean([int(r.get("gpt-4o_judge_match", False)) for r in both_results])

print(f"Number of russian samples: {len(russian_results)}")
print(f"Number of non-russian samples: {len(non_russian_results)}")
print(f"Number of both samples: {len(both_results)}")
print(f"Russian accuracy: {russian_acc:.2%}")
print(f"Non-Russian accuracy: {non_russian_acc:.2%}")
print(f"Both accuracy: {both_acc:.2%}")
```
