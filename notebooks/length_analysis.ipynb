```python
from cresowlve.utils import read_json

gemini_3_pro_results = read_json("../experiments/outputs/gemini-3.1-pro-preview/chgk_en_benchmark_eval_reasoning_model_en_s0_gemini-3.1-pro-preview_20260310_105704.json")
gemini_3_flash_results = read_json("../experiments/outputs/gemini-3-flash-preview/chgk_en_benchmark_eval_reasoning_model_en_s0_gemini-3-flash-preview_20260308_120523.json")
```

```python
gemini_3_pro_thought_lengths = [len(sample["thoughts"]) for sample in gemini_3_pro_results["data"] if sample["thoughts"]]
gemini_3_flash_thought_lengths = [len(sample["thoughts"]) for sample in gemini_3_flash_results["data"] if sample["thoughts"]]
```

```python
import seaborn as sns
import pandas as pd

df = pd.DataFrame({
    "thought_length": gemini_3_pro_thought_lengths + gemini_3_flash_thought_lengths,
    "model": ["Gemini 3 Pro"] * len(gemini_3_pro_thought_lengths) + ["Gemini 3 Flash"] * len(gemini_3_flash_thought_lengths),
})

sns.histplot(data=df, x="thought_length", hue="model", kde=True)
```
